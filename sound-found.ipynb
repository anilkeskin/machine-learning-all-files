```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestCentroid
from sklearn import svm
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from scipy.io import wavfile
import numpy as np
from os.path import dirname, join as pjoin
from scipy.io import wavfile
import scipy.io
import librosa
import os
import pandas as pd
from sklearn.linear_model import LinearRegression
import librosa
a1 = []
a2 = []
test_cases = []
test_cases_names = []
for r, d, files in os.walk("0"):
    for filename in files:
        (data, rate) = librosa.load(os.path.join(r, filename), sr=44100) #sr=NONE
        a1.append(data)
        a2.append(0) 
for r, d, files in os.walk("1"):
    for filename in files:
        (data, rate) = librosa.load(os.path.join(r, filename), sr=44100) #sr=NONE
        a1.append(data)
        a2.append(1)
for r, d, files in os.walk("test"):
    for filename in files:
        (data, rate) = librosa.load(os.path.join(r, filename), sr=44100) #sr=NONE
        test_cases.append(data)
        test_cases_names.append(filename)
final = RandomForestClassifier(n_estimators=100,random_state=0, bootstrap=False, min_samples_leaf=1).fit(a1,a2)

dataFrame = pd.DataFrame(columns=["FileName","Class"])
dataFrame["Class"]  = final.predict(test_cases)
dataFrame["FileName"] = test_cases_names
dataFrame.to_csv("submission.csv",index = False)
print(final.predict(test_cases))
```
