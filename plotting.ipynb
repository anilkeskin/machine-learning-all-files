```python
# QUESTION 4 - B
def Funcsum(n) : 
    result = 0
    if (n<0):
        return result
    else :
        for x in range (0,n):
            result = result + (1/2)**x
    return result
xval=[]
yval=[]
for x in range (-5,10):
    yval.append(Funcsum(x))
    xval.append(x)
plt.stem(xval,yval)
plt.show()
```

```python
# QUESTION 5
import pandas as pd
def shortTerm (n,n_df) :
    result = 0
    for x in range(n-5,n):
        result =result + n_df[x]
    return result/5
def middleTerm (n,n_df) :
    result = 0
    for x in range(n-20,n):
        result =result + n_df[x]
    return result/20
def longTerm (n,n_df) :
    result = 0
    bitis=n-50
    if n==49 : bitis=n-49
    for x in range(bitis,n):
        result =result + n_df[x]
    return result/50
df = pd.read_csv("AAPL.csv")
df1=df["Close"].values
df2=df["Date"].values
df3=df1[49:]
i=49
IsitBuy=0
dates=[]
pre=[]
sayi=0
for val in df3:
    short = shortTerm(i,df1)
    middle = middleTerm(i,df1)
    long = longTerm(i,df1)
    if middle > long and val > short and IsitBuy==0:
        print("Bought ! â> Date : ", df2[i])
        IsitBuy=1
        dates.append(df2[i])
        pre.append(1)
    elif middle < long and val < short and IsitBuy==1:
        print("Sold ! â> Date : ", df2[i])
        IsitBuy=0
        dates.append(df2[i])
        pre.append(-1)
    else :
        dates.append(df2[i])
        pre.append(0)
    i=i+1
veriler = {'Date': dates,'Prediction': pre}
dfcvs = pd.DataFrame(veriler)
dfcvs.to_csv (r'output.csv', index = False, header=True)

#print (dfcvs)
```
