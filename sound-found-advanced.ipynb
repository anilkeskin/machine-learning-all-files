```python
#Q2 1
from pylab import *
def H1(w):
        return 8.0 / ( (1j * w)**3 + 4*((1j*w)**2) + 8*(1j*w) +8 )
f = logspace(-2,2)
#20log|H(ω)|
plot(f, 20*log10(abs(H1(2*pi*f)))); xscale('log')
#Low Pass Filter
```

```python
#Q2 2
def H2(w):
        return  (1j*w)**3 / ( (1j * w)**3 + 2*((1j*w)**2) + 2*(1j*w) +1 )
f = logspace(-2,2)
#20log|H(ω)|
plot(f, 20*log10(abs(H2(2*pi*f)))); xscale('log')
#High Pass filter
```

```python
#Q2 3
import math

def H3(w):
        return ( 1 + 3*(e**(-1j*w)) + 3*(e**(-1j*2*w)) + e**(-1j*3*w) ) / (6 + 2*(e**(-1j*w)))
f = logspace(-2,0)
#20log|H(ω)|
plot(f, 20*log10(abs(H3(2*pi*f)))); xscale('log')
#Low Pass filter
```
